```python
# To load dataset

filename = '/content/drive/MyDrive/CODING_DOJO/DATA_SCIENCE_BOOTCAMP/(02)WEEK2-Pandas/titanic.csv'
df = pd.read_csv(filename)

df
```

```python
# ITEM 1: Display the head of the dataframe.

df.head()
```

```python
# ITEM 2: What is the percentage of people who survived?

average_survival = df.Survived.mean()
percentage_survival = average_survival*100
two_decimals = round(percentage_survival, 2)

print('The percentage of people who survived is ' + str(two_decimals) + '%.')
```

```python
# ITEM 3: How many of each sex survived?

gender_survival = df.groupby(['Sex'])['Survived'].count()

gender_display = gender_survival.reset_index(name='Number of Survivors')

print('''

Based on the data below, there are a total of 314 female survivors and 577 male survivors.

''')
gender_display
```

```python
# ITEM 4-A: Considering only the group of passengers who paid less than $10, what percentage of this group survived?

# To know total number of passengers who paid less than $10
fare_filter = df['Fare'] < 10

# To know total number of passengers who survived
survival_filter = df['Survived'] == 1

# To know total number of passengers who survived and paid less than $10
under10_survivors = df.loc[fare_filter & survival_filter, :]


print('''
Based on the data below, there were 67 passengers who survived and paid less
than $10 for their fare.


''')

under10_survivors
```

```python
# ITEM 4-B: What percentage of the previous group survived?

# To calculate percentage of survivors who paid under $10, 
## we must divide the number of survivors who paid less than $10 by
### the total number of passengers and then multiply the result by 100.

# This is the total number of passengers

survivors_under10 = 67
total_passengers = df['Name'].count()

# This is the formula for getting the percentage of survivors

percentage_u10_survivors = (67 / total_passengers) * 100

print('''
The group of passengers who paid less than $10 and survived represented ''')
print(round(percentage_u10_survivors, 2), '% of the total number of passengers who paid a fare under $10.')
```

```python
# ITEM 5: Overall, what is the average age of those who didn't survive?

non_survivors = df['Survived'] == 0
average_age = df.loc[non_survivors, 'Age'].mean()

print('''
The average age of those who did not survive is''')
print(round(average_age, 2), 'years old.')
```

```python
# ITEM 6: Overall, what is the average age of those who did survive?

survivors = df['Survived'] == 1
average_age = df.loc[survivors, 'Age'].mean()

print('''
The average age of those who survived is''')
print(round(average_age, 2), 'years old.')
```

```python
# ITEM 7: What is the average age of those who did and didn't survive grouped by Sex?

age_survival_average = df.groupby(['Sex', 'Survived'])['Age'].mean()

age_survival_average_display = age_survival_average.reset_index(name='Average Age')


print('''

Based on the data below, the average age of female survivors was 28.85 and the
average age of females who did not survive was 25.05.

Meanwhile, the average age of male survivors was 27.28, while the average 
age of men who did not survive was 31.62 years old.

''')


age_survival_average_display
```
