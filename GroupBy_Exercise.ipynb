```python
# To import pandas library
import pandas as pd
```

```python
# To load data from a url

## To let interpreter know that there is no header, add header = None



df = pd.read_csv('https://raw.githubusercontent.com/mGalarnyk/Tutorial_Data/master/Pandas/CalIt2.data', header = None)

df.head()
```

```python
# INSTRUCTIONS: 
# To rename column 0 to 'Flow_ID'
# To rename column 1 to 'Date'
# To rename column 2 to 'Time'
# To rename column 3 to 'Count'

new_columns = ['Flow_ID', 'Date', 'Time', 'Count']
df.columns = new_columns
df.head()
```

```python
# INSTRUCTIONS: Select all data for the date July 24 2005 having flow id=7

date_filter = df['Date'] == '07/24/05'
flow_id_filter = df['Flow_ID'] == 7
july = df.loc[date_filter & flow_id_filter, :]

print('''

These are the data for July 24, 2005 with Flow Id of 7:

''')
july
```

```python
# INSTRUCTIONS: From the data selected above, sort the result on Count in descending order and output the top 10 rows. Assign the output to a variable named top10.

top10 = july.sort_values('Count', ascending=False).head(10)
july.groupby(["Flow_ID", "Date", "Time", "Count"])



print('''

These are the top 10 rows of the Count column in descending order:

''')

top10
```

```python
# INSTRUCTIONS: For the 10 rows outputted above, use Pandas Apply function to subtract lowest value of the 10 from all of them.

top10.loc[:, 'Count'] = df['Count'].apply(lambda x: x - 1)

print('''

These are the new data with the lowest value from the Count column 
subtracted from the rest of the values in the said column.
The lowest value is 1.

''')

top10
```

```python
# INSTRUCTIONS: Find the average value of the resulting counts


print('''
The average value of the resulting counts is
''')

average = top10.Count.mean()
print(average)
```
