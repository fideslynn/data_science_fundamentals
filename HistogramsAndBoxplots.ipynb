## Creating Graphs

### Histograms

#### Histogram of the Percentage of Dog Owners

```python
plt.style.use('seaborn')

dog_histogram = df.hist(column='Percentage of Dog Owners', 
                        edgecolor='black', 
                        color='#0062FF',
                        range = (0,100))
plt.xlabel("Percentage of Dog Owners in the USA")
plt.ylabel("Count")
plt.show()
```

#### Histogram of the Percentage of Cat Owners

```python
plt.style.use('seaborn')

cat_histogram = df.hist(column='Percentage of Cat Owners', 
                        edgecolor='black', 
                        color='#AE4FDE',
                        range = (0,100))
plt.xlabel("Percentage of Cat Owners in the USA")
plt.ylabel("Count")
plt.show()
```

```python
dog = df['Percentage of Dog Owners']
cat = df['Percentage of Cat Owners']

print('''
PERCENTAGE OF DOG OWNERS
''')
print('MEAN: ', dog.mean())
print('MEDIAN: ', dog.median())
print('MODE: ', dog.mode())
print('MIN: ', dog.min())
print('MAX: ', dog.max())


print('''
----------------------------------------------
''')

print('''
PERCENTAGE OF CAT OWNERS
''')
print('MEAN: ', cat.mean())
print('MEDIAN: ', cat.median())
print('MODE: ', cat.mode())
print('MIN: ', cat.min())
print('MAX: ', cat.max())


```

##### Observations

**What can we see by comparing these two histograms? What information does this tell us?**

*On Dog Ownership*

We first transformed the range to (0, 100) because the data we are using in the graphs represent percentages and they must show the full range between 0% to 100%.

We also tried to determine the mean, median and mode to check if the histogram is reflecting the right picture based on the numbers of these 3.

Without changing the range to go up to 100, the Dog Histogram seems to be left skewed. But when looking at the numbers, it doesn't make sense that it's left skewed because we can see that the mean is greater than the median and the mode (with the mode having the least value). 

For left skewed histograms, what's usually the case is that the mean is less than the median, which is often less than the mode. 

That said, when we transformed the range, we see a more accurate picture. We can affirm that what the graph is showing is consistent with the printed values of the mean, median and mode. 

In short, we can say that the Dog Histogram's shape is right skewed. The mean is larger than the median and the mode. We can therefore say that the skewness value is positive. This means that the values are nearer the 0% point and peaks near the first third of the graph. 

We can deduce from this that owning a dog in the US is not usually the norm as it doesn't reach the middle of the graph. Only about 1/3 of the population choose to be dog owners.


*On Cat Ownership*

The Cat Ownership histogram is also right skewed. It also peaks at around the first third portion of the histrogram. This also shows that 1/3 of the total population sampled choose to be cat owners. 




#### Plotting the Histograms on the Same Plot

```python
plt.style.use('seaborn')

dog = df['Percentage of Dog Owners']
cat = df['Percentage of Cat Owners']

plt.hist(dog,
         bins = 7,
         alpha = 0.72, 
         edgecolor = 'black', 
         color = '#0062FF')

plt.hist(cat, 
         alpha = 0.72, 
         bins = 15,
         edgecolor = 'black', 
         color = '#AE4FDE')

plt.legend(['Percentage of Dog Owners in the USA', 'Percentage of Cat Owners in the USA'], loc='upper left')
plt.xlabel("Percentage of Cat and Dog Owners in the USA")
plt.ylabel("Count")

plt.show()
```

### Boxplots

#### Two boxplots on the same plot: One of the mean number of cats per household and the other of the mean number of dogs

```python
# To set variables

cats_mean = df['Mean Number of Cats']
dogs_mean = df['Mean Number of Dogs per household']
```

```python
# plt.subplots() is a function that returns a tuple containing a figure and axes object
## using fig, axes = plt.subplots() unpacks a tuple into the variables fig and axes
### fig is useful if you want to change figure-level attributes or save the figure as an image file later 

fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (8,5))

# To change boxplot appearance

boxplots = axes.boxplot([cats_mean, dogs_mean],
           notch = True, # If False (default), produces a rectangular box plot. If True, will produce a notched box plot
           labels = ['Cats', 'Dogs'],
           widths = .6,
           patch_artist = True, # When True: produces boxes with the Patch artist; When False: produces boxes with the Line2D artist 
           
           medianprops = dict(linestyle='-', # If provided, this will set the plotting style of the medians.
                              linewidth=2, 
                              color='Orange'),
                        
           boxprops = dict(linestyle='--', # If provided, it will set the plotting style of the boxes.
                           linewidth=2, 
                           color='Black', 
                           facecolor = 'blue', 
                           alpha = .45)
          );

# To change fill color of boxplots

boxplot1 = boxplots['boxes'][0]
boxplot1.set_facecolor('#029AC7')

boxplot2 = boxplots['boxes'][1]
boxplot2.set_facecolor('#00C752')

# To label x and y axes

plt.xlabel('Mean Number of Animals Per Household', fontsize = 15);
plt.ylabel('Mean Number of Animals Per State', fontsize = 15);

# To get or set the current tick locations and labels of the x and y axes

plt.xticks(fontsize = 12);
plt.yticks(fontsize = 12);
```

##### Observations

**What can we see by comparing these two boxplots? What information does this tell us?**

First, we can see that in both boxes, the median is not in the middle. This confirms that the distribution is skewed as we discussed earlier.

On one hand, we see that in the Cats box, the distribution is positively skewed because the median is closer to the bottom. On the other hand, the Dogs box shows that the median is closer towards the top so the distribution is negatively skewed. This is as far as distribution of data is concerned.

There are no outliers in our data.

The notches of the boxes do not overlap. This means that their medians are proven to be statistically different.

This plot shows that cat ownership holds the highest percentage value in animal ownership in US households compared to dog ownership. This is confirmed by the max value of Cat ownership, which is 49.5, which higher compared to Dog ownership (47.9).

However, when looking at the histograms, we can see that there are more dog owners in the US than cat owners on the average. This is confirmed by a higher mean and median value for Dog ownership compared to Cat ownership.
